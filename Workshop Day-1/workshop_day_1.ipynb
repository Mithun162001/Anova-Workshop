Binning/ Discretization Using Pandas

```python
data={'Title':['The Hobbit', 'Robinson Crusoe', 'Moby Dick'],
    'Author':['J.R.R. Tolkien', 'Daniel Defoe', 'Herman Melville']}
df3= pd.DataFrame(data)
df3
```

```python
df3['new col'] = ['age', '3d', 'df']
```

```python
df3['Age'] = [3,13,29]         # Created a new column named new_col and assigned values as 10,20,30
df3['bins'] = pd.cut(x=df3['Age'], bins = [0,10,20,30])    # df['new_bin_column_name']=pd.cut(x=df['column_to_be_binned'], bins=[range of values])
df3
```

```python
# Create New Column of of decade With Labels 10s, 20s, and 30s
df3['decade']= pd.cut(x=df3['Age'], bins=[0,10,20,30], labels=['10s', '20s', '30s'])
df3
```
